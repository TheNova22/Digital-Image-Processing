```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = Image.open('images/comet.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
threshMat = []
thresh = 180
for arr in imgMat:
    threshMat.append(list(map(lambda x :255 if x >= thresh else 0, arr)))
threshMat = np.array(threshMat, dtype=np.uint8)
```

```python
final = Image.fromarray(threshMat)
```

```python
final.save('images/gamma.jpeg')
```
