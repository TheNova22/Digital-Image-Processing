```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = Image.open('images/comet.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
logMat = []
c = 120
print(c)
for arr in imgMat:
    logMat.append(list(map(lambda x :c * np.log(1 + (x/255)), arr)))
logMat = np.array(logMat, dtype=np.uint8)
```

```python
final = Image.fromarray(logMat)
```

```python
final.save('images/log.jpeg')
```
