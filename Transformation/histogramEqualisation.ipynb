```python
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('images/histo.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
n, bins, patches = plt.hist(imgMat.flatten(),range=(0,256),bins=256)
plt.show()
```

```python
normalized = n / sum(n)
```

```python
cSum = np.cumsum(normalized)
```

```python
equaliser = np.array(255 * cSum,dtype=np.uint8)
```

```python
equaliser
```

```python
equalisedImage = []
for i in range(len(imgMat)):
    arr = []
    for j in range(len(imgMat[0])):
        arr.append(equaliser[imgMat[i][j]])
    equalisedImage.append(arr)
```

```python
final = Image.fromarray(np.array(equalisedImage))
```

```python
final.save('images/finHisto.jpeg')
```
