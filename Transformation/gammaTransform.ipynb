```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = Image.open('images/cat.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
gamMat = []
c = 255
gamma = 2
for arr in imgMat:
    gamMat.append(list(map(lambda x :c * ((x/255) ** gamma), arr)))
gamMat = np.array(gamMat, dtype=np.uint8)
```

```python
final = Image.fromarray(gamMat)
```

```python
final.save('images/gamma.jpeg')
```
