```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = Image.open('images/comet.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
negMat = []
for arr in imgMat:
    negMat.append(list(map(lambda x :int( 255 - x), arr)))
negMat = np.array(negMat, dtype=np.uint8)
```

```python
final = Image.fromarray(negMat)
```

```python
final.save('images/nega.jpeg')
```
