```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
img = Image.open('images/comet.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
lap_kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]]) 
```

```python
lapMat = imgMat.copy()
for i in range(1,len(imgMat) - 1):
    for j in range(1,len(imgMat[0]) - 1):
        value = lap_kernel * imgMat[(i - 1):(i + 2), (j - 1):(j + 2)]
        lapMat[i][j] = min(max(0,value.sum()),255)
```

```python
fin = Image.fromarray(lapMat)
```

```python
fin
```

```python
fin.save('images/lap1.jpeg')
```
