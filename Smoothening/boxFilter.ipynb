```python
import pandas as pd
import numpy as np
from PIL import Image
```

```python
n = 3
boxF = [[1] * n for i in range(n)]
```

```python
img = Image.open('images/cat.jpeg').convert('L').resize((640 // 2,426 // 2))
img
```

```python
imgMat = np.array(img)
imgMat
```

```python
def nineMul(i,j,mat):
    diff = [x for x in range(-n,n+1)]
    ct = 0
    prod = 0
    for dx in diff:
        if (i + dx) >= 0 and (i + dx) < len(mat):
            for dy in diff:
                if (j + dy) >= 0 and (j + dy) < len(mat[0]):
                    ct += 1
                    prod += mat[i + dx][j + dy]
    return prod // ct
```

```python
boxxed = imgMat
for i in range(len(imgMat)):
    for j in range(len(imgMat[0])):
        boxxed[i][j] = nineMul(i,j,boxxed)
        # print(boxxed[i][j],imgMat[i][j])
```

```python
final = Image.fromarray(boxxed)
```
